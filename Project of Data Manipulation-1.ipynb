## The over 3000 price by the Type:

```python
top = df[(["Company"] == "all") | (df["Price_euros"] > 3000)]
```

```python
plt.figure(figsize=(10,5), dpi= 150)
sns.histplot(
   top.sort_values('Price_euros').reset_index(),
    y="Company",
    
    hue="TypeName",
    multiple="stack",
    alpha=1,
    
    
)
plt.title("The over 3000 price by the Type ");




## the most company was prduce the gaimng type for over 3000 was dell ,,and for the work staion type the hp was the most one
##also the lenovo was the only one have the notebook for over 3000 
#the razer company has the most expensive laptop .
```

## The most type was produce:

```python
r = top.TypeName.value_counts()
plt.figure(figsize=(10,6))
labels = ['Netbook', 'Notebook' , 'Ultrabook' , 'Gaming'  ]
r.plot.pie(autopct='%.1f%%', labels=None, colors = ['lightskyblue', 'lightpink' , 'lightblue' ,'lightgreen' ]).set_title('The most type produce', fontsize=16)
plt.legend(labels, loc=(1,1));


#•	the most type was produce is netbook by over 60%..and the second type was notebook by 21% and last 2 type was gaming and Ultra book by
```

## Products that weight less than one and the price.

```python
# this Bar Chart shows Products that weight less than one and the price.
plt.figure(figsize=(10,5), dpi= 150)
sns.histplot(
   Weight.sort_values('Price_euros').reset_index(),
    y="Product",
    x='Price_euros',
    hue="Weight",
    multiple="stack",
    alpha=0.8,
    
    
)
plt.title("Products that weight less than one and the price");
```
